## Day 2 (Activation Functions)

* Understand the role of activation functions in introducing non-linearity to neural networks and implement common ones (sigmoid, ReLU, tanh) in Python.

* Activation functions enable neural networks to model complex, non-linear patterns, essential for tasks like image classification or NLP. Without them, networks would be limited to linear transformations.



### Activation functions transform a neuron’s weighted sum ($ z = \mathbf{w} \cdot \mathbf{x} + b $) into an output, introducing non-linearity to model complex patterns.

1. Sigmoid: $ \sigma(z) = \frac{1}{1 + e^{-z}} $, outputs (0,1), used for binary classification but prone to vanishing gradients.

2. ReLU: $ f(z) = \max(0, z) $, outputs [0,∞), fast and avoids vanishing gradients, widely used in deep networks.

3. Tanh: $ \tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} $, outputs (-1,1), centers data, better than sigmoid for hidden layers.

Non-linearity allows networks to approximate arbitrary functions, as per the Universal Approximation Theorem, critical for tasks like cat/dog classification.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    """Compute the sigmoid function."""
    return 1/ (1+ np.exp(-z))

def relu(z):
    """Computes the ReLU function."""
    return np.maximum(0,z)
def tanh(z):
    """Computes the tanh function. (e^z - e^-z)/ (e^z + e^-z)"""
    return np.tanh(z)
```

```python
z= np.linspace(-10, 10, 100)
```

```python
sigmoid_output= sigmoid(z)
relu_output= relu(z)
tanh_output= tanh(z)
```

```python
plt.figure(figsize= (12, 8))
plt.plot(z, sigmoid_output, label='Sigmoid', color='blue')
plt.plot(z, relu_output, label='ReLU', color='green')
plt.plot(z, tanh_output, label='Tanh', color='red')
plt.title('Activation Functions')
plt.xlabel('Input (z)')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
